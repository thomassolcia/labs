# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
########## PLOT DAS IMAGENS ###########

imagem_color1 = Image.open('/content/drive/My Drive/lab02/bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem_color2 = Image.open('/content/drive/My Drive/lab02/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_color3 = Image.open('/content/drive/My Drive/lab02/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')

plt.figure(figsize=(20,4))

plt.subplot(131)
plt.title('Imagem 1')
plt.axis('off')
plt.imshow(imagem_color1)

plt.subplot(132)
plt.title('Imagem 2')
plt.axis('off')
plt.imshow(imagem_color2)

plt.subplot(133)
plt.title('Imagem 3')
plt.axis('off')
plt.imshow(imagem_color3)
plt.show()

########## PLOT DOS HISTOGRAMAS ###########

imagem_color_arr1 = np.array(imagem_color1)
imagem_color_arr2 = np.array(imagem_color2)
imagem_color_arr3 = np.array(imagem_color3)

plt.figure(figsize=(19,4))

red1 = imagem_color_arr1[:,:,0]
red2 = imagem_color_arr2[:,:,0]
red3 = imagem_color_arr3[:,:,0]

plt.subplot(131)
plt.ylim(top=8e6)
plt.hist(red1.ravel(), color='red')
plt.subplot(132)
plt.ylim(top=8e6)
plt.hist(red2.ravel(), color='red')
plt.subplot(133)
plt.ylim(top=8e6)
plt.hist(red3.ravel(), color='red')
plt.ylim(top=8e6)
plt.figure(figsize=(19,4))

green1 = imagem_color_arr1[:,:,1]
green2 = imagem_color_arr2[:,:,1]
green3 = imagem_color_arr3[:,:,1]

plt.subplot(131)
plt.ylim(top=8e6)
plt.hist(green1.ravel(), color='green')
plt.subplot(132)
plt.ylim(top=8e6)
plt.hist(green2.ravel(), color='green')
plt.subplot(133)
plt.ylim(top=8e6)
plt.hist(green3.ravel(), color='green')
plt.ylim(top=8e6)
plt.figure(figsize=(19,4))

blue1 = imagem_color_arr1[:,:,2]
blue2 = imagem_color_arr2[:,:,2]
blue3 = imagem_color_arr3[:,:,2]

plt.subplot(131)
plt.hist(blue1.ravel(), color='blue')
plt.subplot(132)
plt.ylim(top=8e6)
plt.hist(blue2.ravel(), color='blue')
plt.subplot(133)
plt.ylim(top=8e6)
plt.hist(blue3.ravel(), color='blue')
plt.ylim(top=8e6)
```

Olhando para os gráficos gerados, discorram comparando os resultados.



> É possível analizar a partir dos gráficos quais tons são mais frequentes nas imagens, e também em qual proporção elas aparecem. Na imagem 1 temos o azul como cor predominante, na imagem 2 o tom predominante é o verde tendo o azul em quantidade porém mais fraco e na imagem 3 a cor predominante é a vermelha possuindo tons em azul que na imagem em sí quase não são perceptíveis.



# 2a parte

Escolha uma das imagens acima, para continuarmos os experimentos. 

Vamos gerar algumas imagens preto e branco, utilizando limiares diferentes:

*  45
*  100
*  145
*  200
*  245

Escreva o código no espaço abaixo.

```
### Tons de Cinza ###
imagem3_arr = np.array(imagem_color3)

red = imagem3_arr[:,:,0]
green = imagem3_arr[:,:,1]
blue = imagem3_arr[:,:,2]

imagem3_gray = np.mean(imagem3_arr, axis=2)

### 45 ###
imagem3_arr45 = np.array(imagem_color3)

red = imagem3_arr45[:,:,0]
green = imagem3_arr45[:,:,1]
blue = imagem3_arr45[:,:,2]

imagem3_gray45 = np.mean(imagem3_arr45, axis=2)

imagem3_arr45[imagem3_gray45 >= 45] = 255
imagem3_arr45[imagem3_gray45 < 45] = 0

### 100 ###
imagem3_arr100 = np.array(imagem_color3)

red = imagem3_arr100[:,:,0]
green = imagem3_arr100[:,:,1]
blue = imagem3_arr100[:,:,2]

imagem3_gray100 = np.mean(imagem3_arr100, axis=2)

imagem3_arr100[imagem3_gray100 >= 100] = 255
imagem3_arr100[imagem3_gray100 < 100] = 0

### 145 ###
imagem3_arr145 = np.array(imagem_color3)

red = imagem3_arr145[:,:,0]
green = imagem3_arr145[:,:,1]
blue = imagem3_arr145[:,:,2]

imagem3_gray145 = np.mean(imagem3_arr145, axis=2)

imagem3_arr145[imagem3_gray145 >= 145] = 255
imagem3_arr145[imagem3_gray145 < 145] = 0

### 200 ###
imagem3_arr200 = np.array(imagem_color3)

red = imagem3_arr200[:,:,0]
green = imagem3_arr200[:,:,1]
blue = imagem3_arr200[:,:,2]

imagem3_gray200 = np.mean(imagem3_arr200, axis=2)

imagem3_arr200[imagem3_gray200 >= 200] = 255
imagem3_arr200[imagem3_gray200 < 200] = 0

### 245 ###
imagem3_arr245 = np.array(imagem_color3)

red = imagem3_arr245[:,:,0]
green = imagem3_arr245[:,:,1]
blue = imagem3_arr245[:,:,2]

imagem3_gray245 = np.mean(imagem3_arr245, axis=2)

imagem3_arr245[imagem3_gray245 >= 245] = 255
imagem3_arr245[imagem3_gray245 < 245] = 0
```

Plote as imagens geradas. (Escreva o código no espaço abaixo)

```
plt.figure(figsize=(20,14))

plt.subplot(331)
plt.title('Original')
plt.axis('off')
plt.imshow(imagem_color3)

plt.subplot(332)
plt.title('Tons de Cinza')
plt.axis('off')
plt.imshow(Image.fromarray(imagem3_gray))

plt.subplot(333)
plt.title('Limiar 45')
plt.axis('off')
plt.imshow(imagem3_arr45)

plt.subplot(334)
plt.title('Limiar 100')
plt.axis('off')
plt.imshow(imagem3_arr100)

plt.subplot(335)
plt.title('Limiar 145')
plt.axis('off')
plt.imshow(imagem3_arr145)

plt.subplot(336)
plt.title('Limiar 200')
plt.axis('off')
plt.imshow(imagem3_arr200)

plt.subplot(337)
plt.title('Limiar 245')
plt.axis('off')
plt.imshow(imagem3_arr245)
```

O que você consegue extrair deste experimento? Escreva no espaço abaixo.

> É possível notar que com o aumento dos limiares as imagens vão ficando mais escuras. Isso por que quanto mais próximo do valor de 255 os tons que antes eram cinzas são 'setados' para 255, logo deixando a imagem mais escura.
